```python
train_video_dir = "data/videos/train/"
val_video_dir = "data/videos/validation/"

train_frames_dir = "data/frames/train/"
val_frames_dir = "data/frames/validation/"

print(len(os.listdir(train_video_dir)))
print(len(os.listdir(val_video_dir)))

print(len(os.listdir(train_frames_dir)))
print(len(os.listdir(val_frames_dir)))
```

### Utils

```python
def exctract_frames(video_name, videos_dir, frames_dir, frame_interval=6):

  print(f"Processing video: {video_name}")
  video_path = os.path.join(videos_dir, video_name)
  frames_subdir = os.path.join(frames_dir, video_name.split(".")[0])

  if not os.path.exists(frames_subdir):
    os.makedirs(frames_subdir)
    print(f"Created new frames_subdir: {frames_subdir}")
  else:
    print(f"Already exists: {frames_subdir}")

    if len(os.listdir(frames_subdir)) > 0:
      print(f"There are {len(os.listdir(frames_subdir))} files, so skipping this video.")
      return 0


  vidcap = cv2.VideoCapture(video_path)
  success,image = vidcap.read()

  count = 0

  pbar = tqdm(total=1500)  # approx calculation: ~5min = 300 sec = 300 x 30frames = 9K frames / 6 = 1500 saved frames 

  while success:
    frame_name = f"frame_{count}.jpg"
    frame_path = os.path.join(frames_subdir, frame_name)

    if count % frame_interval == 0:
      cv2.imwrite(frame_path, image)     # save every 6th frames as in the paper
      pbar.update(1)

    success,image = vidcap.read()
    count += 1

  pbar.close()
  
  print("Done\n")
```

### Main

```python
# train videos

for i, video_name in enumerate(sorted(os.listdir(train_video_dir))):
  print(f"{i + 1} / {len(os.listdir(train_video_dir))} {video_name}\n")

  status = exctract_frames(video_name, videos_dir=train_video_dir, frames_dir=train_frames_dir)

  print("-" * 50 + "\n")
```

```python
# validation videos

for i, video_name in enumerate(sorted(os.listdir(val_video_dir))):
  print(f"{i + 1} / {len(os.listdir(val_video_dir))} {video_name}\n")

  status = exctract_frames(video_name, videos_dir=val_video_dir, frames_dir=val_frames_dir)

  print("-" * 50 + "\n")
```
